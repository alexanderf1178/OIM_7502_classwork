```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # optional may be helpful for plotting percentage
import numpy as np
import pandas as pd
import seaborn as sb  # optional to set plot theme
import yfinance as yf  # yfinance library for stock data

sb.set_theme()  # optional to set plot theme

# Default start and end dates
DEFAULT_START = dt.date.isoformat(dt.date.today() - dt.timedelta(365))
DEFAULT_END = dt.date.isoformat(dt.date.today())

class Stock:
    def __init__(self, symbol, start=DEFAULT_START, end=DEFAULT_END):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.data = self.get_data()

    def get_data(self):
        """Method that downloads data and stores in a DataFrame"""
        stock_data = yf.download(self.symbol, start=self.start, end=self.end)
        stock_data.index = pd.to_datetime(stock_data.index)
        self.calc_returns(stock_data)
        return stock_data

    def calc_returns(self, df):
        """Method that adds change and return columns to data"""
        df['Change'] = df['Close'].diff()
        df['Return'] = df['Close'].pct_change() * 100

    def plot_return_dist(self):
        """Method that plots instantaneous returns as histogram"""
        if 'Return' not in self.data.columns:
            print("Return data not available. Please calculate returns first.")
            return
        
        plt.figure(figsize=(10, 6))
        sb.histplot(self.data['Return'].dropna(), bins=50, kde=True)
        plt.title(f'Return Distribution for {self.symbol}')
        plt.xlabel('Daily Return (%)')
        plt.ylabel('Frequency')
        plt.show()

    def plot_performance(self):
        """Method that plots stock object performance as percent"""
        plt.figure(figsize=(12, 8))
        initial_price = self.data['Close'][0]
        self.data['Performance'] = (self.data['Close'] / initial_price - 1) * 100
        plt.plot(self.data.index, self.data['Performance'], label='Performance')
        plt.title(f'{self.symbol} Stock Performance')
        plt.xlabel('Date')
        plt.ylabel('Performance (%)')
        plt.legend()
        plt.show()

def main():
    # Example usage for Pfizer stock
    test = Stock(symbol='PFE')  # Create an instance of the Stock class with 'PFE' symbol
    print(test.data)  # Print the downloaded data
    test.plot_performance()  # Plot the stock performance
    test.plot_return_dist()  # Plot the return distribution

if __name__ == '__main__':
    main()
```

```python
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick  # optional may be helpful for plotting percentage
import numpy as np
import pandas as pd
import seaborn as sb  # optional to set plot theme
import yfinance as yf  # yfinance library for stock data

sb.set_theme()  # optional to set plot theme

# Default start and end dates
DEFAULT_START = dt.date.isoformat(dt.date.today() - dt.timedelta(365))
DEFAULT_END = dt.date.isoformat(dt.date.today())

class Stock:
    def __init__(self, symbol, start=DEFAULT_START, end=DEFAULT_END):
        self.symbol = symbol
        self.start = start
        self.end = end
        self.data = self.get_data()

    def get_data(self):
        stock_data = yf.download(self.symbol, start=self.start, end=self.end)
        stock_data.index = pd.to_datetime(stock_data.index)
        print(stock_data.head())  # Print the first few rows to verify data
        print(stock_data.columns)  # Print column names to verify
        self.calc_returns(stock_data)
        return stock_data

    def calc_returns(self, df):
        df['Change'] = df['Close'].diff()
        df['Return'] = df['Close'].pct_change() * 100

    def plot_return_dist(self):
        if 'Return' not in self.data.columns:
            print("Return data not available. Please calculate returns first.")
            return
        
        plt.figure(figsize=(10, 6))
        sb.histplot(self.data['Return'].dropna(), bins=50, kde=True)
        plt.title(f'Return Distribution for {self.symbol}')
        plt.xlabel('Daily Return (%)')
        plt.ylabel('Frequency')
        plt.show()

    def plot_performance(self):
        if self.data.empty:
            print("Data is empty. Cannot plot performance.") #check to see if data is missing
            return
        
        plt.figure(figsize=(12, 8))
        initial_price = self.data['Close'].iloc[0]
        self.data['Performance'] = (self.data['Close'] / initial_price - 1) * 100
        plt.plot(self.data.index, self.data['Performance'], label='Performance')
        plt.title(f'{self.symbol} Stock Performance')
        plt.xlabel('Date')
        plt.ylabel('Performance (%)')
        plt.legend()
        plt.show()

def main():
    test = Stock(symbol='PFE')
    print(test.data)
    test.plot_performance()
    test.plot_return_dist()

if __name__ == '__main__':
    main()
```
